```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.common.alert import Alert


def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(5)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error")


def product_page(driver):
    try:
        product = driver.find_element(By.CSS_SELECTOR, 'a[href="/products"]' )
        product.click()
        print("product page clicked")
    except NoSuchElementException:
        print("Error")


def verify_product_page(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='All Products']" )
        print("ALL products page verified")
    except NoSuchElementException:
        print("Error")


def add_cart(driver):
    try:
        for product_id in range (1,3):  # 1 and 2
            
            add_cart = driver.find_element(By.XPATH, f"//a[@data-product-id='{product_id}']")
            add_cart.click()
            print(f" Item {product_id} added")
            time.sleep(5)
            

            continue_shopping = driver.find_element(By.XPATH, '//button[normalize-space() = "Continue Shopping"]')
            continue_shopping.click()
            print("continue shopping clicked")
            time.sleep(2)

    except NoSuchElementException:
        print("Error")

def cart(driver):
    try:
        cart = driver.find_element(By.CSS_SELECTOR, 'a[href = "/view_cart"]')
        cart.click()
        print("cart clicked")
        time.sleep(5)
        
        driver.find_element(By.ID, "product-1")
        driver.find_element(By.ID, "product-2")
        print("Both products are added to Cart")
        

    except NoSuchElementException:
        print("Error")

def verify_cart(driver):
    try:
        for product_id in range (1,3):
            row = driver.find_element(By.ID, f"product-{product_id}")

            price = row.find_element(By.CLASS_NAME , "cart_price" ).text
            quantity = row.find_element(By.CLASS_NAME, "cart_quantity").text
            total = row.find_element(By.CLASS_NAME, "cart_total_price").text

            print(f"Product {product_id} verification:")
            print(f"  Price    : {price}")
            print(f"  Quantity : {quantity}")
            print(f"  Total    : {total}")
            print("-"*30)


        print("Price, Quantity and Total verified for both products")

    except NoSuchElementException:
        print("Error")


def main():
    driver = Launch_automation_exercise()
    time.sleep(5)

    try:
        verify_home_page(driver)
        time.sleep(5)
        
        product_page(driver)
        time.sleep(5)
        
        verify_product_page(driver)
        time.sleep(5)

        add_cart(driver)
        time.sleep(5)

        cart(driver)
        time.sleep(5)

        verify_cart(driver)
        time.sleep(5)

    finally:
        driver.quit()
        print("Browser closed")

if __name__ == "__main__":
    main()
    
        
```
