```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.common.alert import Alert


def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error")


def product_page(driver):
    try:
        product = driver.find_element(By.CSS_SELECTOR, 'a[href="/products"]' )
        product.click()
        print("product page clicked")
    except NoSuchElementException:
        print("Error")


def verify_product_page(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='All Products']" )
        print("ALL products page verified")
    except NoSuchElementException:
        print("Error")


def product_details(driver):
    try:
        view_product = driver.find_element(By.CSS_SELECTOR,'a[href="/product_details/1"]' )
        view_product.click()
        print("view product page loaded")
        
    except NoSuchElementException:
        print("Error")

def verify_details(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='Blue Top']")
        print("product name verified")
        time.sleep(2)

        driver.find_element(By.XPATH,  "//p[contains(normalize-space(), 'Category') and contains(normalize-space(), 'Women')]") 
        print("category verified")
        time.sleep(2)

        driver.find_element(By.XPATH, "//span[contains(normalize-space(),'Rs.')]")
        print("price verified")
        time.sleep(2)
        
        driver.find_element(By.XPATH, "//b[text() = 'Availability:']")
        print("Availability verified")
        time.sleep(2)
        
        driver.find_element(By.XPATH, "//b[text() = 'Condition:']")
        print("Condition verified")
        time.sleep(2)
        
        driver.find_element(By.XPATH, "//b[text() = 'Brand:']")
        print("Brand verified")
        time.sleep(2)

    except NoSuchElementException:
        print("Error")


def main():
    
    driver = Launch_automation_exercise()
    time.sleep(5)

    try:
        verify_home_page(driver)
        time.sleep(5)
    
        product_page(driver)
        time.sleep(5)
    
        verify_product_page(driver)
        time.sleep(5)
    
        product_details(driver)
        time.sleep(5)

        verify_details(driver)
        time.sleep(5)

    finally:
        driver.quit()
        print("Browser closed")

if __name__ == "__main__":
    main()
    
```
