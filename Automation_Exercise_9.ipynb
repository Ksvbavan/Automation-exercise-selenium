```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.common.alert import Alert


def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error")


def product_page(driver):
    try:
        product = driver.find_element(By.CSS_SELECTOR, 'a[href="/products"]' )
        product.click()
        print("product page clicked")
    except NoSuchElementException:
        print("Error")


def verify_product_page(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='All Products']" )
        print("ALL products page verified")
    except NoSuchElementException:
        print("Error")

def search_product(driver, product_name):
    try:
        search_product = driver.find_element(By.ID , "search_product")
        search_product.send_keys(product_name )
        print("product name entered")
        time.sleep(2)

        search_button = driver.find_element(By.ID, "submit_search")
        search_button.click()
        print("search option clicked")
        time.sleep(2)
        
    except NoSuchElementException:
        print("Error")


def verify_search_product(driver, product_name):
    try:
        driver.find_element(By.XPATH, f"//p[normalize-space()='{product_name}']")
        print("searched product result verified")

    except NoSuchElementException:
        print("Error")
        
def main():

    product_name ="Blue Top"
    driver = Launch_automation_exercise()
    time.sleep(5)

    try:
        verify_home_page(driver)
        time.sleep(5)
    
        product_page(driver)
        time.sleep(5)
    
        verify_product_page(driver)
        time.sleep(5)
        
        search_product(driver, product_name)
        time.sleep(5)
        
        verify_search_product(driver, product_name)
        time.sleep(5)

    finally:
        driver.quit()
        print("Browser closed")

if __name__ == "__main__":
    main()
    
```
