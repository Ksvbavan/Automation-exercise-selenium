```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.common.alert import Alert


def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error finding Homepage")

def contact_us(driver):
    try:
        contact_us = driver.find_element(By.CSS_SELECTOR, 'a[href = "/contact_us"]')
        contact_us.click()
        print("contact us page clicked")
    except NoSuchElementException:
       print("Error message NOT displayed")


def get_in_touch_page(driver):
    try:
        driver.find_element(By.XPATH, "//h2[text() = 'Get In Touch']")
        print("Get in touch page verified")
    except NoSuchElementException:
        print("Error")

def details(driver, username, email_text, subject_text, message_text, file_path):
    try:
        name = driver.find_element(By.CSS_SELECTOR, '[data-qa="name"]')
        name.send_keys(username)
        print("name entered")
        time.sleep(2)
        
        email = driver.find_element(By.CSS_SELECTOR, '[data-qa="email"]')
        email.send_keys(email_text)
        print("email entered")
        time.sleep(2)

        subject = driver.find_element(By.CSS_SELECTOR, '[data-qa="subject"]')
        subject.send_keys(subject_text)
        print("subject entered")
        time.sleep(2)
        
        message = driver.find_element(By.CSS_SELECTOR, '[data-qa="message"]')
        message.send_keys(message_text)
        print("message entered")
        time.sleep(2)

        upload_file = driver.find_element(By.NAME, "upload_file")
        upload_file.send_keys(file_path)
        print("file Uploaded")
        time.sleep(2)
        
        
    except NoSuchElementException:
        print("Error")

def submit_button(driver):
    try:
        submit_button = driver.find_element(By.CSS_SELECTOR, '[data-qa ="submit-button"]')
        submit_button.click()
        print("submit button clicked")
        time.sleep(5)

        
        alert = driver.switch_to.alert
        alert.accept()
        print("popup ok is submitted")
        
    except NoSuchElementException:
        print("Error")

def verify_success_message(driver):
    try:
        driver.find_element(By.CSS_SELECTOR, "div.status.alert.alert-success")
        print("Success! Your details have been submitted successfully")
    except NoSuchElementException:
        print("Error")

def home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        Home.click()
        print("Homepage clicked and verified")
    except NoSuchElementException:
        print("Error finding Homepage")

def main():

    username = "Bavan Kumar V"
    email = "ksvbavan99@gmail.com"
    password = "12345678901234567890"
    subject = "Greetings"
    message = "Hello world"
    file_path =(r"C:\Users\Bavan Kumar\Desktop\AUTOMATION\hello-world-pixel-7680x4320-15168.png")

   
    driver = Launch_automation_exercise()
    time.sleep(5)

    verify_home_page(driver)
    time.sleep(5)
    
    contact_us(driver)
    time.sleep(5)
    
    get_in_touch_page(driver)
    time.sleep(5)
    
    details(driver, username, email, subject, message, file_path)
    time.sleep(5)
    
    submit_button(driver)
    time.sleep(5)
    
    verify_success_message(driver)
    time.sleep(5)

    home_page(driver)
    time.sleep(5)
    
    driver.quit()

if __name__ == "__main__":
    main()
```
