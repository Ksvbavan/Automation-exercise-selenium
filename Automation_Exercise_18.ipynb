```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time


def launch_automation_exercise():
    url = "https://automationexercise.com"
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"

    driver = webdriver.Chrome(options=options)
    driver.implicitly_wait(5)
    driver.get(url)

    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        driver.find_element(By.CSS_SELECTOR, "a[href='/']")
        print("Home page verified")
    except NoSuchElementException:
        print("Home page not found")


def open_products_page(driver):
    try:
        driver.find_element(By.CSS_SELECTOR, "a[href='/products']").click()
        print("Products page opened")
    except NoSuchElementException:
        print("Products link not found")


def verify_products_page(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='All Products']")
        print("All Products page verified")
    except NoSuchElementException:
        print("All Products heading not found")

def verify_category_title(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space() = 'Category']")
        print("category title verified")
    except NoSuchElementException:
        print("error")

def select_women(driver):
    try:
        women = driver.find_element(By.CSS_SELECTOR, "a[href='#Women']")
        women.click()
        print("women option clicked")
        time.sleep(2)

        dress = driver.find_element(By.CSS_SELECTOR, "a[href='/category_products/1']")
        dress.click()
        print("dress clicked")
        time.sleep(2)
        
    except NoSuchElementException:
        print("error")

def verify_women_dress(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='Women - Dress Products']")
        print("Women - Dress Products verified")

    except NoSuchElementException:
        print("error")

def select_men(driver):
    try:
        men = driver.find_element(By.CSS_SELECTOR, "a[href='#Men']")
        men.click()
        print("men option clicked")
        time.sleep(2)

        t_shirts = driver.find_element(By.CSS_SELECTOR, "a[href='/category_products/3']")
        t_shirts.click()
        print("t shirt clicked")
        time.sleep(2)
        
    except NoSuchElementException:
        print("error")

        
def verify_men_dress(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='Men - Tshirts Products']")
        print("Men - Tshirts Products verified")

    except NoSuchElementException:
        print("error")
        
def main():
    driver = launch_automation_exercise()
    time.sleep(3)

    try:
        verify_home_page(driver)
        time.sleep(5)

        open_products_page(driver)
        time.sleep(5)

        verify_products_page(driver)
        time.sleep(5)

        verify_category_title(driver)
        time.sleep(5)

        select_women(driver)
        time.sleep(5)
        
        verify_women_dress(driver)
        time.sleep(5)

        select_men(driver)
        time.sleep(5)

        verify_men_dress(driver)
        time.sleep(5)

    finally:
        driver.quit()
        print("Browser closed")


if __name__ == "__main__":
    main()
```
