```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time

def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error finding Homepage")


def subscriptions(driver):
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        print("Page scrolled")
        time.sleep(2)
        
        driver.find_element(By.XPATH, "//h2[normalize-space()='Subscription']")
        print("Subscription page visible")

    except NoSuchElementException:
        print("Error")
        
def scrollup(driver):
    try:
        scrollup = driver.find_element(By.ID, "scrollUp")
        scrollup.click()
        print("page scrolled")
        
    except NoSuchElementException:
        print("Error")

def automation_text(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space() ='Full-Fledged practice website for Automation Engineers']" )
        print("Full-Fledged practice website for Automation Engineers (Text verified)")
        
    except NoSuchElementException:
        print("Error")


def main():

    driver = Launch_automation_exercise()
    time.sleep(5)

    try:
        verify_home_page(driver)
        time.sleep(5)
        
        subscriptions(driver)
        time.sleep(5)
        
        scrollup(driver)
        time.sleep(5)
        
        automation_text(driver)
        time.sleep(5)

        
    
        
    finally:
        
        driver.quit()
        print("Browser closed")

if __name__ == "__main__":
    main()
```
