```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time

def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error finding Homepage")

def signup_login(driver):
    try:
        signup = driver.find_element(By.CSS_SELECTOR, 'a[href="/login"]')
        signup.click()
        print("signup/login clicked")

    except NoSuchElementException:
        print("Error")

def verify_login_account(driver):
    try:
        New_user = driver.find_element(By.XPATH, "//h2[text() = 'Login to your account']")
        print("login Page Verified")
    except NoSuchElementException:
        print("Error")

def login_to_your_account(driver, email, password_value):
    try:
        login_email = driver.find_element(By.CSS_SELECTOR, '[data-qa = "login-email"]')
        login_email.send_keys(email)
        print("login email entered")
        time.sleep(2)

        login_password = driver.find_element(By.CSS_SELECTOR, '[data-qa = "login-password"]')
        login_password.send_keys(password_value)
        print("login password entered")
        time.sleep(2)

        login_button = driver.find_element(By.CSS_SELECTOR, '[data-qa = login-button]')
        login_button.click()
        time.sleep(2)

    except NoSuchElementException:
        print("Error")

def verify_login_error_message(driver):
    try:
        error_msg = driver.find_element(
            By.XPATH, "//p[text()='Your email or password is incorrect!']")
        print("Your email or password is incorrect!")

    except NoSuchElementException:
       print("⚠️ Error message NOT displayed")

def main():

    username = "Bavan Kumar V"
    email = "ksvbavan99@gmail.com"
    password = "12345678901234567890"
    incorrect_email = "vbavan@gmail.com"
    incorrect_password = "1234567890"
   
    driver = Launch_automation_exercise()
    time.sleep(5)

    verify_home_page(driver)
    time.sleep(5)

    signup_login(driver)
    time.sleep(5)

    verify_login_account(driver)
    time.sleep(5)

    login_to_your_account(driver, incorrect_email, incorrect_password)
    time.sleep(5)

    verify_login_error_message(driver)
    time.sleep(5)
    
    driver.quit()

if __name__ == "__main__":
    main()
```
