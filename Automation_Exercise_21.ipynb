```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.common.alert import Alert


def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error")


def product_page(driver):
    try:
        product = driver.find_element(By.CSS_SELECTOR, 'a[href="/products"]' )
        product.click()
        print("product page clicked")
    except NoSuchElementException:
        print("Error")


def verify_product_page(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='All Products']" )
        print("ALL products page verified")
    except NoSuchElementException:
        print("Error")

def safe_click(driver, element):
    driver.execute_script("arguments[0].scrollIntoView({block:'center'});", element)
    time.sleep(1)
    driver.execute_script("arguments[0].click();", element)

def product_details(driver):
    try:
        view_product = driver.find_element(By.CSS_SELECTOR,'a[href="/product_details/1"]' )
        safe_click(driver, view_product)
        print("view product page loaded")
        
    except NoSuchElementException:
        print("Error")


def review(driver):
    try:
        name_input = driver.find_element(By.ID , "name")
        name_input.send_keys("Bavan")
        print("name entered")
        time.sleep(2)

        email_input = driver.find_element(By.ID, "email")
        email_input.send_keys("Bavan@123.com")
        print("email entered")
        time.sleep(2)

        review_input = driver.find_element(By.ID, "review")
        review_input.send_keys("greetings")
        print("review entered")
        time.sleep(2)

        submit = driver.find_element(By.ID, "button-review")
        submit.click()
        print("review submitted")
        
    except NoSuchElementException:
        print("Error")


def main():
    
    driver = Launch_automation_exercise()
    time.sleep(5)

    try:
        verify_home_page(driver)
        time.sleep(5)
    
        product_page(driver)
        time.sleep(5)
    
        verify_product_page(driver)
        time.sleep(5)
    
        product_details(driver)
        time.sleep(5)

        review(driver)
        time.sleep(5)


    finally:
        driver.quit()
        print("Browser closed")

if __name__ == "__main__":
    main()
    
```
