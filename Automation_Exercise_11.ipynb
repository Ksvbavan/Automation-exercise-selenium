```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.common.alert import Alert


def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error")


def cart(driver):
    try:
        cart = driver.find_element(By.CSS_SELECTOR, 'a[href = "/view_cart"]')
        cart.click()
        print("cart clicked")
        time.sleep(5)

    except NoSuchElementException:
        print("Error")

        
def subscriptions(driver):
    try:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        print("Page scrolled")
        time.sleep(2)
        
        driver.find_element(By.XPATH, "//h2[normalize-space()='Subscription']")
        print("Subscription page visible")

    except NoSuchElementException:
        print("Error")


def enter_email(driver, email):
    try:
        enter_email = driver.find_element(By.ID , "susbscribe_email")
        enter_email.send_keys(email)
        print("email entered")
        time.sleep(2)

        submit_button = driver.find_element(By.ID, "subscribe")
        submit_button.click()
        print("submit option clicked")
        time.sleep(2)
        
    except NoSuchElementException:
        print("Error")
        
def main():

    email = "123@gmail.com"
    driver = Launch_automation_exercise()
    time.sleep(5)

    try:
        verify_home_page(driver)
        time.sleep(5)

        cart(driver)
        time.sleep(5)
    
        subscriptions(driver)
        time.sleep(5)
        
        enter_email(driver, email)
        time.sleep(5)

    finally:
        driver.quit()
        print("Browser closed")

if __name__ == "__main__":
    main()
    
        
```
