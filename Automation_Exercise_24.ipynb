```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.common.alert import Alert


def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error")

def product_page(driver):
    try:
        product = driver.find_element(By.CSS_SELECTOR, 'a[href="/products"]' )
        product.click()
        print("product page clicked")
    except NoSuchElementException:
        print("Error")


def verify_product_page(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='All Products']" )
        print("ALL products page verified")
    except NoSuchElementException:
        print("Error")


def add_cart(driver):
    try:
        for product_id in range (1,3):  # 1 and 2
            
            add_cart = driver.find_element(By.XPATH, f"//a[@data-product-id='{product_id}']")
            add_cart.click()
            print(f" Item {product_id} added")
            time.sleep(5)
            

            continue_shopping = driver.find_element(By.XPATH, '//button[normalize-space() = "Continue Shopping"]')
            continue_shopping.click()
            print("continue shopping clicked")
            time.sleep(2)

    except NoSuchElementException:
        print("Error")


def cart(driver):
    try:
        cart = driver.find_element(By.CSS_SELECTOR, 'a[href = "/view_cart"]')
        cart.click()
        print("cart clicked")
        time.sleep(5)
        
        driver.find_element(By.ID, "product-1")
        driver.find_element(By.ID, "product-2")
        print("Both products are added to Cart")
        

    except NoSuchElementException:
        print("Error")



def proceed_to_checkout(driver):
    try:
        check_out = driver.find_element(By.XPATH, '//a[normalize-space ()= ("Proceed To Checkout")]')
        check_out.click()
        print("proceed to check out clicked")
    except NoSuchElementException:
        print("Error")

def register_login(driver):
    try:
        register_login = driver.find_element(By.XPATH, '//a[normalize-space() = "Register / Login"]')
        register_login.click()
        print("register/login clicked")
    except NoSuchElementException:
        print("Error")

def verify_new_user_page(driver):
    try:
        New_user = driver.find_element(By.XPATH, "//h2[text() = 'New User Signup!']")
        print("New User Page Verified")

    except NoSuchElementException:
        print("Error")


def signup_name_email(driver, username, email):
    try:
        signup_name = driver.find_element(By.CSS_SELECTOR, '[data-qa="signup-name"]')
        signup_name.send_keys(username)
        print("Name Entered")
        signup_email = driver.find_element(By.CSS_SELECTOR, '[data-qa ="signup-email"]')
        signup_email.send_keys(email)
        print("Email Entered")
        signup_submit = driver.find_element(By.CSS_SELECTOR, '[data-qa ="signup-button"]')
        signup_submit.click()
        print("Signup clicked")
    except NoSuchElementException:
        print("Error")


def verify_account_info(driver):
    try:
        verify_account_info = driver.find_element(By.XPATH, "//b[text() = 'Enter Account Information']")
        print("Verified Enter Account info page")
    except NoSuchElementException:
        print("Error")

def fill_details(driver, password_value):
    try:
        title = driver.find_element(By.ID, "id_gender1")
        title.click()
        print("Title Selected")
        time.sleep(2)
        password = driver.find_element(By.CSS_SELECTOR, '[data-qa ="password"]')
        password.send_keys(password_value)
        print("Password Entered")
        days = driver.find_element(By.ID, "days")
        days.send_keys("10")
        print("Day Selected")
        months = driver.find_element(By.ID, "months")
        months.send_keys("February")
        print("month Selected")
        year = driver.find_element(By.ID, "years")
        year.send_keys("1999")
        print("year Selected")

    except NoSuchElementException:
        print("Error")

def select_checkbox (driver):
    try:
        checkbox1 = driver.find_element(By.ID, "newsletter")
        checkbox1.click()
        print("newsletter checkbox selected")
        
        checkbox2 = driver.find_element(By.ID, "optin")
        checkbox2.click()
        print("special offers checkbox selected")
        
    except NoSuchElementException:
        print("Error")

def fill_details_2(driver):
    try:
        first_name = driver.find_element(By.ID, "first_name")
        first_name.send_keys("Bavan")
        print('First Name Entered')
        time.sleep(2)
        
        last_name = driver.find_element(By.ID, "last_name")
        last_name.send_keys("Kumar V")
        print('Last Name Entered')
        time.sleep(2)
        
        company = driver.find_element(By.ID, "company")
        company.send_keys("Amazon")
        print("Company Name Entered")
        time.sleep(2)
        
        address = driver.find_element(By.ID, "address1")
        address.send_keys("65, KTR Estate Vijayapuram")
        print("Address Entered")
        time.sleep(2)

        country = driver.find_element(By.ID, "country")
        country.send_keys("India")
        print("country name Entered")
        time.sleep(2)

        state = driver.find_element(By.ID, "state")
        state.send_keys("Tamil Nadu")
        print("state Entered")
        time.sleep(2)

        city = driver.find_element(By.ID, "city")
        city.send_keys("Thiruvarur")
        print("city Entered")
        time.sleep(2)

        zipcode = driver.find_element(By.ID, "zipcode")
        zipcode.send_keys("610001")
        print("zipcode Entered")
        time.sleep(2)

        mobile = driver.find_element(By.ID, "mobile_number")
        mobile.send_keys("9494949494")
        print("mobile number Entered")
        time.sleep(2)

        create_account = driver.find_element(By.CSS_SELECTOR, '[data-qa=create-account]')
        create_account.click()
        print("create account clicked")

    except NoSuchElementException:
        print("Error")

def verify_account_created(driver):
    try:
        driver.find_element(By.XPATH, "//b[text() = 'Account Created!']")
        print("Account Creation Verified")
    except NoSuchElementException:
        print("Error")

def continue_button(driver):
    try:
        continue_button = driver.find_element(By.CSS_SELECTOR, '[data-qa=continue-button]')
        continue_button.click()
        print("continue selected")
    except NoSuchElementException:
        print("Error")

def verify_login_account_username(driver, username):
    try:
        New_user = driver.find_element(By.XPATH, f"//a//b[text() = '{username}']")
        print("Logged in as username")
    except NoSuchElementException:
        print("Error")

def delete_account(driver):
    try:
        delete_account = driver.find_element(By.CSS_SELECTOR, 'a[href="/delete_account"]')
        delete_account.click()
        print("delete account option selected")

    except NoSuchElementException:
        print("Error")


def download_invoice(driver):
    try:
        download_invoice = driver.find_element(By.CSS_SELECTOR, "a[href='/download_invoice/1400']")
        download_invoice.click()
        print("download invoice clicked")
        time.sleep(2)

        continue_button = driver.find_element(By.CSS_SELECTOR, '[data-qa=continue-button]' )
        continue_button.click()
        print("continue clicked")
        
    except NoSuchElementException:
        print("Error")
        

def verify_account_deleted(driver):
    try:
        driver.find_element(By.XPATH, "//b[text() = 'Account Deleted!']")
        print("Account Deletion Verified")
        time.sleep(2)
        button = driver.find_element(By.CSS_SELECTOR, '[data-qa=continue-button]')
        button.click()
        print("continue selected")
    except NoSuchElementException:
        print("Error")




def verify_address(driver, fields):
    try:

        delivery_address = driver.find_element(By.ID, "address_delivery").text
        billing_address =  driver.find_element(By.ID, "address_invoice").text 

        for field in fields:
            assert field in delivery_address
            
        print("Delivery address Verified")
        time.sleep(2)

        for field in fields:
            assert field in billing_address

        print("Billing address Verified")
        time.sleep(2)
    
    except NoSuchElementException:
        print("Error")



def description_and_place_order(driver):
    try:
        message_input = driver.find_element(By.NAME, "message")
        message_input.send_keys("Greetings")
        print("Message entered")
        time.sleep(2)
        
        place_order = driver.find_element(By.CSS_SELECTOR, "a[href='/payment']")
        place_order.click()
        print("place order clicked")
        time.sleep(2)
        
    except NoSuchElementException:
        print("Error")

        
def payment(driver):

    try:
        name_on_card_input = driver.find_element(By.NAME, "name_on_card")
        name_on_card_input.send_keys("Bavan")
        time.sleep(2)
    
        card_number_input = driver.find_element(By.NAME, "card_number")
        card_number_input.send_keys("1234 5678 9012 3456")
        time.sleep(2)
    
        cvc_input = driver.find_element(By.NAME, "cvc")
        cvc_input.send_keys("123")
        time.sleep(2)
    
        expiry_month_input = driver.find_element(By.NAME, "expiry_month")
        expiry_month_input.send_keys("12")
        time.sleep(2)
        
        expiry_year_input = driver.find_element(By.NAME, "expiry_year")
        expiry_year_input.send_keys("12")
        time.sleep(2)
    
        print("Payment details entered")

        pay_and_confirm = driver.find_element(By.ID, "submit")
        pay_and_confirm.click()

        print("pay and confirm button clicked")
        
        
    except NoSuchElementException:
        print("Error")


def verify_order_confirmation(driver):

    try:

        driver.find_element(By.XPATH, "//p[normalize-space() = 'Congratulations! Your order has been confirmed!']")
        print("order succesfully placed :)")
        time.sleep(2)
        
    except NoSuchElementException:
        print("Error")



def main():


    username = "Bavan Kumar V"
    email = "ksvbavan99@gmail.com"
    password = "12345678901234567890"
    first_name = "Bavan"
    last_name = "Kumar V"
    full_name = f"{first_name} {last_name}"
    company = "Amazon"
    address1 = "65, KTR Estate Vijayapuram"
    country = "India"
    state = "Tamil Nadu"
    city = "Thiruvarur"
    zipcode ="610001"
    mobile_number = "9494949494"
    fields = [full_name, company, address1, country, state, city, zipcode, mobile_number]
   
   
    driver = Launch_automation_exercise()
    time.sleep(5)

    try:
        verify_home_page(driver)
        time.sleep(5)

        
        product_page(driver)
        time.sleep(5)
        
        verify_product_page(driver)
        time.sleep(5)
        
        add_cart(driver)
        time.sleep(5)
        
        cart(driver)
        time.sleep(5)
        
        proceed_to_checkout(driver)
        time.sleep(5)
    
        register_login(driver)
        time.sleep(5)

        verify_new_user_page(driver)
        time.sleep(5)
    
        signup_name_email(driver, username, email)
        time.sleep(5)
    
        verify_account_info(driver)
        time.sleep(5)
    
        fill_details(driver, password)
        time.sleep(5)
    
        select_checkbox (driver)
        time.sleep(5)
    
        fill_details_2(driver)
        time.sleep(5)
    
        verify_account_created(driver)
        time.sleep(5)
    
        continue_button(driver)
        time.sleep(5)
        
        verify_login_account_username(driver, username)
        time.sleep(5)

        cart(driver)
        time.sleep(5)
        
        proceed_to_checkout(driver)
        time.sleep(5)

        
        verify_address(driver, fields)
        time.sleep(5)        
        
        description_and_place_order(driver)
        time.sleep(5)
        
        payment(driver)
        time.sleep(5)
        
        verify_order_confirmation(driver)
        time.sleep(5)

        download_invoice(driver)
        time.sleep(5)
    
        delete_account(driver)
        time.sleep(5)
    
        verify_account_deleted(driver)
        time.sleep(5)

        

    finally:
        driver.quit()
        print("Browser closed")

if __name__ == "__main__":
    main()
    
```
