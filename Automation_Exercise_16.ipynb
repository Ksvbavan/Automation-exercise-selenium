```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time

def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error finding Homepage")

def signup_login(driver):
    try:
        signup = driver.find_element(By.CSS_SELECTOR, 'a[href="/login"]')
        signup.click()
        print("signup clicked")

    except NoSuchElementException:
        print("Error")
def verify_login_account(driver):
    try:
        New_user = driver.find_element(By.XPATH, "//h2[text() = 'Login to your account']")
        print("New User Page Verified")
    except NoSuchElementException:
        print("Error")

def login_to_your_account(driver, email, password):
    try:
        login_email = driver.find_element(By.CSS_SELECTOR, '[data-qa = "login-email"]')
        login_email.send_keys(email)
        print("login email entered")
        time.sleep(2)

        login_password = driver.find_element(By.CSS_SELECTOR, '[data-qa = "login-password"]')
        login_password.send_keys(password)
        print("login password entered")
        time.sleep(2)

        login_button = driver.find_element(By.CSS_SELECTOR, '[data-qa = login-button]')
        login_button.click()
        print("login success")
        time.sleep(2)
        
    except NoSuchElementException:
        print("Error")


def verify_login_account_username(driver, username):
    try:
        New_user = driver.find_element(By.XPATH, f"//a//b[text() = '{username}']")
        print("Logged in as username")
        
    except NoSuchElementException:
        print("Error")


def product_page(driver):
    try:
        product = driver.find_element(By.CSS_SELECTOR, 'a[href="/products"]' )
        product.click()
        print("product page clicked")
    except NoSuchElementException:
        print("Error")


def verify_product_page(driver):
    try:
        driver.find_element(By.XPATH, "//h2[normalize-space()='All Products']" )
        print("ALL products page verified")
    except NoSuchElementException:
        print("Error")

def safe_click(driver, element):
    driver.execute_script("arguments[0].scrollIntoView({block:'center'});", element)
    time.sleep(1)
    driver.execute_script("arguments[0].click();", element)


def add_cart(driver):
    try:
        for product_id in range (1,3):  # 1 and 2
            
            add_cart = driver.find_element(By.XPATH, f"//a[@data-product-id='{product_id}']")
            safe_click(driver, add_cart)
            print(f" Item {product_id} added")
            time.sleep(5)
            

            continue_shopping = driver.find_element(By.XPATH, '//button[normalize-space() = "Continue Shopping"]')
            safe_click(driver, continue_shopping)
            print("continue shopping clicked")
            time.sleep(2)

    except NoSuchElementException:
        print("Error")

def cart(driver):
    try:
        cart = driver.find_element(By.CSS_SELECTOR, 'a[href = "/view_cart"]')
        safe_click(driver, cart)

        print("cart clicked")
        time.sleep(5)
        
        driver.find_element(By.ID, "product-1")
        driver.find_element(By.ID, "product-2")
        print("Both products are added to Cart")
        

    except NoSuchElementException:
        print("Error")

def proceed_to_checkout(driver):
    try:
        check_out = driver.find_element(By.XPATH, '//a[normalize-space ()= ("Proceed To Checkout")]')
        check_out.click()
        print("proceed to check out clicked")
    except NoSuchElementException:
        print("Error")

def verify_address(driver, fields):
    try:

        delivery_address = driver.find_element(By.ID, "address_delivery").text
        billing_address =  driver.find_element(By.ID, "address_invoice").text 

        for field in fields:
            assert field in delivery_address
            
        print("Delivery address Verified")
        time.sleep(2)

        for field in fields:
            assert field in billing_address

        print("Billing address Verified")
        time.sleep(2)
    
    except NoSuchElementException:
        print("Error")

def description_and_place_order(driver):
    try:
        message_input = driver.find_element(By.NAME, "message")
        message_input.send_keys("Greetings")
        print("Message entered")
        time.sleep(2)
        
        place_order = driver.find_element(By.CSS_SELECTOR, "a[href='/payment']")
        place_order.click()
        print("place order clicked")
        time.sleep(2)
        
    except NoSuchElementException:
        print("Error")

def payment(driver):

    try:
        name_on_card_input = driver.find_element(By.NAME, "name_on_card")
        name_on_card_input.send_keys("Bavan")
        time.sleep(2)
    
        card_number_input = driver.find_element(By.NAME, "card_number")
        card_number_input.send_keys("1234 5678 9012 3456")
        time.sleep(2)
    
        cvc_input = driver.find_element(By.NAME, "cvc")
        cvc_input.send_keys("123")
        time.sleep(2)
    
        expiry_month_input = driver.find_element(By.NAME, "expiry_month")
        expiry_month_input.send_keys("12")
        time.sleep(2)
        
        expiry_year_input = driver.find_element(By.NAME, "expiry_year")
        expiry_year_input.send_keys("12")
        time.sleep(2)
    
        print("Payment details entered")

        pay_and_confirm = driver.find_element(By.ID, "submit")
        pay_and_confirm.click()

        print("pay and confirm button clicked")
        
        
    except NoSuchElementException:
        print("Error")

def verify_order_confirmation(driver):

    try:

        driver.find_element(By.XPATH, "//p[normalize-space() = 'Congratulations! Your order has been confirmed!']")
        print("order succesfully placed :)")
        time.sleep(2)
        
    except NoSuchElementException:
        print("Error")

def delete_account(driver):
    try:
        delete_account = driver.find_element(By.CSS_SELECTOR, 'a[href="/delete_account"]')
        delete_account.click()
        print("delete account option selected")

    except NoSuchElementException:
        print("Error")

def verify_account_deleted(driver):
    try:
        driver.find_element(By.XPATH, "//b[text() = 'Account Deleted!']")
        print("Account Deletion Verified")
        time.sleep(2)
        button = driver.find_element(By.CSS_SELECTOR, '[data-qa=continue-button]')
        button.click()
        print("continue selected")
    except NoSuchElementException:
        print("Error")

def main():

    username = "Bavan Kumar V"
    email = "ksvbavan99@gmail.com"
    password = "12345678901234567890"
    first_name = "Bavan"
    last_name = "Kumar V"
    full_name = f"{first_name} {last_name}"
    company = "Amazon"
    address1 = "65, KTR Estate Vijayapuram"
    country = "India"
    state = "Tamil Nadu"
    city = "Thiruvarur"
    zipcode ="610001"
    mobile_number = "9494949494"
    fields = [full_name, company, address1, country, state, city, zipcode, mobile_number]
   
    driver = Launch_automation_exercise()
    time.sleep(5)

    try:
        verify_home_page(driver)
        time.sleep(5)
    
        signup_login(driver)
        time.sleep(5)

        verify_login_account(driver)
        time.sleep(5)


        login_to_your_account(driver, email, password)
        time.sleep(5)

        verify_login_account_username(driver, username)
        time.sleep(5)
        
        product_page(driver)
        time.sleep(5)
    
        verify_product_page(driver)
        time.sleep(5)

        add_cart(driver)
        time.sleep(5)

        cart(driver)
        time.sleep(5)
        
        proceed_to_checkout(driver)
        time.sleep(5)

        verify_address(driver, fields)
        time.sleep(5)
        
        description_and_place_order(driver)
        time.sleep(5)

        payment(driver)
        time.sleep(5)
        
        verify_order_confirmation(driver)
        time.sleep(5)
        
        delete_account(driver)
        time.sleep(5)
        
        verify_account_deleted(driver)
        time.sleep(5)

        
   
        
    finally:
        
        driver.quit()
        print("Browser closed")

if __name__ == "__main__":
    main()
```
