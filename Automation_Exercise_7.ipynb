```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time
from selenium.webdriver.common.alert import Alert


def Launch_automation_exercise():
    url = 'https://automationexercise.com'
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    options.page_load_strategy = "eager"
    driver = webdriver.Chrome(options = options)
    driver.implicitly_wait(1)
    driver.get(url)
    print("Browser launched")
    return driver


def verify_home_page(driver):
    try:
        Home = driver.find_element(By.CSS_SELECTOR, 'a[href="/"]')
        print("Homepage Verified")
    except NoSuchElementException:
        print("Error finding Homepage")


def test_case(driver):
    try:
        test_case = driver.find_element(By.XPATH, "//button[normalize-space() ='Test Cases'] ")
        test_case.click()
        print("Test case submitted")

    except NoSuchElementException:
        print("Error finding Homepage")

def verify_test_case(driver):
    try:
        driver.find_element(By.XPATH, "//b[text() = 'Test Cases']")
        print("Test case page verified succesfully")
        
    except NoSuchElementException:
        print("Error finding Homepage")


def main():

   
    driver = Launch_automation_exercise()
    time.sleep(5)

    verify_home_page(driver)
    time.sleep(5)

    test_case(driver)
    time.sleep(5)

    verify_test_case(driver)
    time.sleep(5)
    
    driver.quit()

if __name__ == "__main__":
    main()
    
```
